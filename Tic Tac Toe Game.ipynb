# Tic Tac Toe Game

```python
def printBoard(board):
    # Print the current state of the Tic Tac Toe board
    print(f"{board[0]} | {board[1]} | {board[2]} ")
    print(f"--|---|---")
    print(f"{board[3]} | {board[4]} | {board[5]} ")
    print(f"--|---|---")
    print(f"{board[6]} | {board[7]} | {board[8]} ")

def checkWin(board):
    wins = [[0, 1, 2], [3, 4, 5], [6, 7, 8], [0, 3, 6], [1, 4, 7], [2, 5, 8], [0, 4, 8], [2, 4, 6]]
    for win in wins:
        if board[win[0]] == board[win[1]] == board[win[2]]:
            if board[win[0]] == 'X':
                # X player has won the match
                print("X Won the match")
                return 1
            elif board[win[0]] == 'O':
                # O player has won the match
                print("O Won the match")
                return 0
    return -1 

if __name__ == "__main__":
    board = ['0', '1', '2', '3', '4', '5', '6', '7', '8']
    turn = 'X'
    print("Welcome to Tic Tac Toe")
    moves = 0  # To count the number of moves
    while moves < 9:
        printBoard(board)
        print(f"{turn}'s Chance")
        value = int(input("Please enter a value: "))
        while not (0 <= value <= 8) or board[value] == 'X' or board[value] == 'O':
            print("Invalid move. Try again.")
            value = int(input("Please enter a value: "))

        board[value] = turn
        moves += 1

        cwin = checkWin(board)
        if cwin != -1:
            print("Match over")
            break

        if moves == 9:
            print("It's a draw")
            break

        turn = 'O' if turn == 'X' else 'X'
```
